```python
df2 = df[((df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
# Remove (USA) from event name
df2.loc[:, 'Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# Clean up athlete age
df2.loc[:, 'athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# Remove h from athlete performance
df2.loc[:,'Athlete performance'] =  df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# Drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category
df2 = df2.drop(['Athlete club', 'Athlete country','Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2 = df2.rename(columns = {
'Year of event':'year',
'Event dates':'race_day',
'Event name':'race_name',
'Event distance/length':'race_length',
'Event number of finishers':'race_number_of_finishers',
'Athlete performance':'athlete_performance',
'Athlete gender':'athlete_gender',
'Athlete average speed':'athlete_average_speed',
'Athlete ID':'athlete_id'
})
```

```python
# Charts and graphs
```

```python
sns.histplot(df3['race_length']);
```

Through the graph we can se that Short races (50km) will attract more participants than long races (50mi)

```python
sns.histplot(df3, x = 'race_length', hue = 'athlete_gender');
```

Through the graph we can se that Short races (50km) will attract more participants than long races (50mi); Ratio of male and female who participate races is same.

```python
sns.displot(df3[df3['race_length'] == '50mi']['athlete_average_speed']);
```

We can see that the average speed of athlete are allocation from 2 to 16; Speeds from 6 to 7 are achieved by most people

```python
sns.violinplot(data = df3, x='race_length', y='athlete_average_speed', hue='athlete_gender', split = True, inner = 'quart', linewidth=1 );
```

- Male athletes generally have slightly higher average speeds compared to female athletes, but the distributions are fairly close.
- The violin plot helps illustrate that the variation in speed is similar between genders and across race lengths.
- There are some high-speed outliers, particularly among male athletes.

```python
sns.lmplot(data = df3, x='athlete_age', y='athlete_average_speed', hue='athlete_gender');
```

- Athlete performance generally declines with age.
- Males tend to have slightly higher speeds than females.
- The majority of athletes fall within the 20-50 age range.
- Younger athletes exhibit more variability in speed, with some outliers having exceptionally high performance.

```python
# Seasons for the data -> Slower in sumber than winter?
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2

# split between two decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter' )
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=False)
```
